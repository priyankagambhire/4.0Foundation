```python
def bar_plot(variable):
    var=df[variable]
    varValue=var.value_counts()
    mpt.figure(figsize=(9,3))
    mpt.bar(varValue.index,varValue)
    mpt.xticks(varValue.index,varValue.index.values)
    mpt.ylabel("Frequency")
    mpt.title(variable)
    mpt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
def bar_plot(variable):
    var=df[variable]
    varValue=var.value_counts()
    mpt.figure(figsize=(9,3))
    colors = sns.color_palette(palette='pastel',n_colors=4)
    mpt.bar(varValue.index,varValue,color=colors)
    mpt.xticks(varValue.index,varValue.index.values)
    mpt.ylabel("Frequency")
    mpt.title(variable)
    mpt.show()
    print("{}: \n {}".format(variable,varValue))

columns_plot=['GarageSpaces','YearBuilt']
for i  in columns_plot:
    bar_plot(i)
```
